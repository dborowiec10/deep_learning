```python
import numpy as np
import matplotlib.pyplot as plt
import plot_vec
```

```python
# function to plot multiple vectors
orange = '#FF9A13'
blue = '#1190FF'

def plotVectors(vecs, cols, alpha=1):
    """
    Plot set of vectors.

    Parameters
    ----------
    vecs : array-like
        Coordinates of the vectors to plot. Each vectors is in an array. For
        instance: [[1, 3], [2, 2]] can be used to plot 2 vectors.
    cols : array-like
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
    alpha : float
        Opacity of vectors

    Returns:

    fig : instance of matplotlib.figure.Figure
        The figure of the vectors
    """
    plt.figure()
    plt.axvline(x=0, color='#A9A9A9', zorder=0)
    plt.axhline(y=0, color='#A9A9A9', zorder=0)

    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                   alpha=alpha)
```

```python
# equation 
# Ax = b

# can have
# 0 solutions 
# 1 solution = lines intersect
# infinite solutions = lines interposed
```

```python
# d and e are 2 solutions to the system Ax = b
# Ad = b
# Ae = b
```

```python
# potential solution is: Az = b where z = ad + (1 - a)e for any value of a
# Az = A(ad + (1 - a)e)
#    = Ada + Ae(1 - a)
```

```python
# we also know that
# Ad = Ae = b
# so...
# Az = ba + b(1 - a)
#    = ba + b - ba
#    = b
# So Az = b and z is also a solution
```

```python
# If there are more equations than unknows the system is called overdetermined
```

```python
# If there is more unknowns than equations the system is called underdetermined
```

```python
# lets examine equations: 
# y = 2x + 1
# y = 6x - 2
```

```python
x = np.arange(-10, 10)
y = 2*x + 1
y1 = 6*x - 2
```

```python
plt.figure()
plt.plot(x, y)
plt.plot(x, y1)
plt.xlim(-2, 10)
plt.ylim(-2, 10)
# draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')
```

```python
# On the above figure, the two lines are crossing so there is one unique solution
# If they are superimposed there are a infinite number of solutions since each points of the lines corresponds to an intersection
# If they are parallel, there is no solution.
```

```python
# weights
a = 2
b = 1

# vector coordinates
u = [0, 0, 1, 3]
v = [2, 6, 2, 1]

plt.quiver([u[0], a*u[0], b*v[0]],
           [u[1], a*u[1], b*v[1]],
           [u[2], a*u[2], b*v[2]],
           [u[3], a*u[3], b*v[3]],
           angles='xy', scale_units='xy', scale=1, color=[orange, orange, blue])
plt.xlim(-1, 8)
plt.ylim(-1, 8)
```

```python
# span of vectors u and v is the set of points that can be reached by changing the combinations of a and b
```

```python
# column figure
```

```python
# It is also possible to represent the set of equations by considering that the solution
# vector b corresponds to a linear combination of each columns multiplied by their weights.
```

```python
# The column representation is
```

```python
# The columns of A give us the directions we can travel by and their weights are the length of the way in that direction.
```

```python
# Variables known are
#     [1/2  -1]
#     [1     1]

# as such the column figure is:

#     [1/2]       [-1]   [-1]
# x * [1  ] + y * [1]  = [4 ]

# goal is to find x and y such that combination of the variables and weights produces the vector [-1, 4]
```

```python
x = np.arange(-10, 10)

y = 0.5 * x + 1
y1 = -x + 4
```

```python
plt.figure()
plt.plot(x, y)
plt.plot(x, y1)
plt.xlim(-2, 10)
plt.ylim(-2, 10)
# draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')
```

```python
# We can see that the solution (the intersection of the lines representing our two equations) is x=2 and y=2.
```
